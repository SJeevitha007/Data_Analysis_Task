```python
import pandas as pd
import numpy as np

df = pd.read_excel('/content/drive/My Drive/Data_Analyst_Task.xls')
df.shape
```

```python
df = df.dropna(subset=["Product Amount with GST"])
```

```python
df.sort_values(by='Lead Registered Time', ascending=False, inplace=True)
df.drop_duplicates(subset=['User ID', 'Product Code', 'Payment Status'], keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
duplicates = df.duplicated(subset=['User ID', 'Product Code'], keep=False)
df[duplicates]
```

```python
df.drop_duplicates(subset=['User ID', 'Product Code'], keep='first', inplace=True)
def replace_initiated_with_paid(row):
    if row['Payment Status'] == 'Initiated':
        prior_rows = df[
            (df['User ID'] == row['User ID']) &
            (df['Product Code'] == row['Product Code']) &
            (df['Lead Registered Time'] < row['Lead Registered Time'])
        ]
        for _, prior_row in prior_rows.iterrows():
            if prior_row['Payment Status'] == 'Paid':
                return 'Paid'
    return row['Payment Status']
df['Payment Status'] = df.apply(replace_initiated_with_paid, axis=1)
df.reset_index(drop=True, inplace=True)
```
